# Principais fatores de rotatividade
* A taxa de rotatividade é alta quando O total_day_minutes está entre 210 e 300 min.
* Quando o total_day_charge (cobranca de ligacoes diurnas) está entre 40 e 50.
* Quando nao tem o servico de number_vmail_messages.
* Quando o account_length (numero de meses que o cliente esta utilizando o provedor) esta entre 60 e 120.
* Quando o total_day_calls (total de ligacoes por dia)esta entre 85 e 115.
* Quando o total_eve_minutes está entre 160 e 260 minutos.
* Quando o total_eve_calls esta entre 90 e 115.
* Quando o total_eve_charge esta entre 15 e 18.
* Quando o total_night_minutes esta entre 190 e 220 minutos.
* Quando o total_nigh_calls esta entre 90 e 110.
* Quando o total_night_charge esta entre 7.5 e 10.
* Quando o total_intl_minutes esta entre 9 e 12.
* Quando o total_intl_calls é 1.
* Quando o total_intl_charge esta entre 2.5 e 3.
* Quando o area_code é = '415' (Canada).
